```python
# Classify the movie based on Ratings 
data.columns 
```

```python
def rating(rating):
    if rating>=7.0:
        return "Excellent"
    elif rating>=5-6:
        return "Good"
    elif rating>=3-5:
        return "Average"
    else:
        return "Poor"
```

```python
data['Rating_Category'] = data['Rating'].apply(rating)
```
